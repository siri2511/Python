### **Arguments Passing**


*   Pass by value
*   Pass by reference



```python
def change_number(num):
    num+=10

def change_list(num_list):
    num_list.append(20)

num_val=10
print("Pass by value")
print("num_val before function call: ",num_val)
change_number(num_val)
print("num_val after function call: ",num_val)

num_list=[10,20,30]
print("Pass by reference")
print("num_list before function call: ",num_list)
change_list(num_list)
print("num_list after function call: ",num_list)
```

### **Argument Types**


*   Positional arguments
*   Keyword arguments
*   Default arguments
*   Variable argument count



```python
def display1(flight_number, seating_capacity):
    print("Flight number: ",flight_number)
    print("Seating capacity: ",seating_capacity)
    print("\n")

print("code-1: positional arguments")
display1("FN789",200)

def display2(flight_number,seating_capacity):
    print("Flight number: ",flight_number)
    print("Seating capacity: ",seating_capacity)
    print("\n")

print("code-2: keyword arguments")
display2(seating_capacity=200,flight_number="FN789")

def display3(flight_number,flight_make="Boeing",seating_capacity=250):
    print("Flight number: ",flight_number)
    print("Flight make: ",flight_make)
    print("Seating capacity: ",seating_capacity)
    print("\n")

print("code-3: default arguments")
display3("FN789","Eagle")

def display4(passenger_name,*baggage_tuple):
    print("Passenger name: ",passenger_name)
    total_wt=0
    for baggage_wt in baggage_tuple:
        total_wt+=baggage_wt
    print("Total baggage weight: ",total_wt)
    print("\n")

print("code-4: variable argument count")
display4("Jack",12,8,16)
```

### **Variable Types**<br>


*   Global variables
*   Local variables



```python
wt_limit = 30

def baggage_check(baggage_wt):
    extra_baggage_charge=0
    if not(baggage_wt>=0 and baggage_wt<=wt_limit):
        extra_baggage_charge=(baggage_wt-wt_limit)*100
    return extra_baggage_charge

def update_baggage_limit(new_wt_limit):
    wt_limit=new_wt_limit
    print("This airline now allows baggage limit till ",wt_limit,"kgs")

print("This airline allows baggage limit till ",wt_limit,"kgs")
print("Pay extra baggage charge of ",baggage_check(35),'rupees')
update_baggage_limit(40)
print("This airline now allows baggage limit till ",wt_limit,"kgs")
```

### **EXERCISE**
1. function factorial(num) returns factorial of a number
2. function is_palindrome(num) checks whether the given number is palindrome or not
3. function check_amicable_numbers(num1,num2) checks whether two numbers are amicable or not <br>
two numbers are amicable if sum of divisors of num1 is equal to num2 and sum of divisors of num2 is equal to num1
4. function right_shift(num,n) returns a given number num roated right by n positions
5. function strong_number(num) checks if the given number is strong or not <br>
a number is strong number if factorial of each digit of the number is equal to the given number

```python
def factorial(num):
    fact =1
    for i in range(1,num+1):
        fact*=i
    return fact

n=int(input("Enter number: "))
print("Factorial of ",n," is ",factorial(n))
```

```python
def is_palindrome(num):
    return num==num[::-1]

n=input("Enter number:")
if is_palindrome(n):
    print(n," is a palindrome")
else:
    print(n," is not a palindrome")
```

```python
def sum_divisors(num):
    sum=0
    for i in range(1,num):
        if num%i==0:
            sum+=i
    return sum

def check_amicable_numbers(num1,num2):
    sum1=sum_divisors(num1)
    sum2 = sum_divisors(num2)
    return sum1==num2 and sum2==num1

a,b= (int(i) for i in input("Enter two numbers").split(","))
if check_amicable_numbers(a,b):
    print(a," ",b," are amicable")
else:
    print(a," ",b," are not amicable")
```

```python
def right_shift(num,n):
    return num>>n

a,b= (int(i) for i in input("Enter two numbers").split(","))
print("The desired number is ",right_shift(a,b))
```

```python
def factorial(num):
    prod=1
    for i in range(1,num+1):
        prod*=i
    return prod

def check_strong_number(num):
    sum=0
    n = num
    while num!=0:
        sum+=factorial(num%10)
        num=num//10
    return sum==n

n=int(input("Enter number:"))
if check_strong_number(n):
    print(n,' is a strong number')
else:
    print(n,' is not a strong number')
```
