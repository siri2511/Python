### **Error Handling**

### **Invoking Stored Functions**

```python
DELIMITER //
CREATE FUNCTION increase_by_100 (number INT) RETURNS INT
BEGIN
    RETURN number+100
END;
//
```

```python
import mysql.connector
con = mysql.connector.connect(host='',user='',password='',database='')

cur=con.cursor()
cur.execute('select increase_by_100(%s)',[10])
print(cur.fetchone())

cur.close()
con.close()
```

### **Invoking Stored Procedures**

```python
DELIMITER //

CREATE PROCEDURE promotion_discount(IN discount INT, INOUT price INT)
BEGIN
SET price=price*(1-discount/100)
END;
//
```

```python
import mysql.connector
con = mysql.connector.connect(host='',user='',password='',database='')

cur=con.cursor()
args=[10,1000]
res = cur.callproc('promotion_discount',args)
print(res)

cur.close()
con.close()
```

## **PYTHON PROGRAMMING USING MONGODB**

```python
from pymongo import MongoClient

#connecting to monogdb database
conn_obj=MongoClient('mongo://hostname:portnumber/')
print(type(conn_obj))

#alternate method for creating connection object
conn_obj=MongoClient(hostname, portnumber)
```

```python
#selecting database
database_obj = conn_obj['database_name']

#alternate method
database_obj = conn_obj.database_name

#if database exists it will be returned else new one will be created
```

```python
#accessing collection
collection_obj = database_obj['collection_name']

#alternate method
collection_obj = database_obj.collection_name

#if collection exists it will be returned else new one will be created
```

```python
#fetching databases list
db_list = conn_obj.list_database_names()
print(db_list)

#fetching list of collections
database_obj = conn_obj['database_name']
col_list = database_obj.list_collection_names()
print(col_list)
```

```python
#inserting single document into collection
database_obj = conn_obj['database_name']
collection_obj = database_obj['collection_name']

document = {'Name':'John','Regd_id':'1','Age':20,'Sex':'M'}

#insert_one returns a result object
returnval = collection_obj.insert_one(document)
#inserted_id: used to retrieve the insertion id of the result object
print(returnval.inserted_id)
```

```python
#insert multiple documents into collection
database_obj = conn_obj['database_name']
collection_obj = database_obj['collection_name']

documents = [{'Name':'Kelley','Regd_id':'2','Age':21,'Sex':'F'},
             {'Name':'Hannah','Regd_id':'3','Age':25,'Sex':'F'},
             {'Name':'Michael','Regd_id':'4','Age':30,'Sex':'M'}
]

returnval = collection_obj.insert_many(documents)
print(returnval.inserted_ids)
```

```python
#retrieving documents from collection
database_obj = conn_obj['database_name']
collection_obj = database_obj['collection_name']

print('Find one document')
doc = collection_obj.find_one()
print(doc)
#output will be of the format: {'_id': ObjectId('document_id'),data}

print("Find all documents")
docs = collection_obj.find()
for doc in docs:
    print(doc)
```

```python
#fetching documents using conditions
database_obj = conn_obj['database_name']
collection_obj = database_obj['collection_name']

#query needs to be passed to find() as its first argument
docs=collection_obj.find({'Sex':'F'})
for doc in docs:
    print(doc)
```

```python
#fetch students whose age is greater than 21
database_obj = conn_obj['database_name']
collection_obj = database_obj['collection_name']

docs = collection_obj.find({'Age':{'$gt':21}})
for doc in docs:
    print(doc)
```

```python
# $regex to filter based on regular expressions
database_obj = conn_obj['database_name']
collection_obj = database_obj['collection_name']

docs = collection_obj.find({'Name':{'$regex':'^R'}})
for doc in docs:
    print(doc)
```

```python
#projecting only select columns
database_obj = conn_obj['database_name']
collection_obj = database_obj['collection_name']

docs = collection_obj.find({},{'_id':0,'Name':1,'Age':1})
for doc in docs:
    print(doc)
```

```python
#sorting
database_obj = conn_obj['database_name']
collection_obj = database_obj['collection_name']

#by default ascending: 1, for descending: -1
docs = collection_obj.find({}).sort('Age',1)
for doc in docs:
    print(doc)
```

```python
#limiting the documents
database_obj = conn_obj['database_name']
collection_obj = database_obj['collection_name']

# no of documents as argument to limit()
docs = collection_obj.find({}).limit(2)
for doc in docs:
    print(doc)
```

```python
#updating documents in a collection\
database_obj = conn_obj['database_name']
collection_obj = database_obj['collection_name']

#updates first match document
#if query criteria is not provided, update_one() modifies the first document
collection_obj.update_one({'Regd_id':'3'},{'$set':{'Age':25}})

#update_many() returns update object
#update object has modified_count attribute: number of updated records
collection_obj.update_many({'Regd_id':{'$gt':'4'}},{'$set':{'Age':24}})

docs = collection_obj.find({},{'_id':0,'Regd_id':1,'Age':1})
for doc in docs:
    print(doc)
```

```python
#deleting documents from collection
database_obj = conn_obj['database_name']
collection_obj = database_obj['collection_name']

#query criteria not provided, delete_one() deletes first document
collection_obj.delete_one({'Regd_id':'1'})

#delete_many() returns deletion object
#delete object has deleted_count attribute: contains number of records deleted
collection_obj.delete_many({'Regd_id':{'$gt':'4'}})

docs = collection_obj.find({})
for doc in docs:
    print(doc)
```
