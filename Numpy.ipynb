### **Operations on Numpy Arrays**

### Accessing Array Elements

```python
#accessing elements from 1d array
cars=np.array(['cheverolet chevelle malibu', 'buick skylark 320', 'plymouth satellite', 'amc rebel sst', 'ford torino'])

#accessing 2nd car from the array
cars[1]
```

```python
#accessing elements from 2d array
car_names=np.array(['cheverolet chevelle malibu', 'buick skylark 320', 'plymouth satellite', 'amc rebel sst', 'ford torino'])
horsepower = [130, 165, 150, 150, 140]

car_hp_arr = np.array([car_names,horsepower])
car_hp_arr
```

```python
#accessing horsepower
car_hp_arr[0]
```

```python
#accessing horsepower
car_hp_arr[1]
```

```python
#accessing second car
car_hp_arr[0,1]
```

```python
#accessing name of the last car
car_hp_arr[0,-1]
```

### Slicing

```python
#slicing from 1d array
cars=np.array(['cheverolet chevelle malibu', 'buick skylark 320', 'plymouth satellite', 'amc rebel sst', 'ford torino'])

cars[1:4]
```

```python
#slicing from 2d array
car_names=np.array(['cheverolet chevelle malibu', 'buick skylark 320', 'plymouth satellite', 'amc rebel sst', 'ford torino'])
horsepower = [130, 165, 150, 150, 140]
acceleration = [18, 15, 18, 16, 17]

car_hp_acc_arr = np.array([car_names, horsepower, acceleration])
car_hp_acc_arr
```

```python
#accessing name and horsepower
car_hp_acc_arr[0:2]
```

```python
#accessing name and horsepower of last two cars
car_hp_acc_arr[0:2, 3:5]
```

```python
#accessing name, horsepower and acceleartion of first 3 cars
car_hp_acc_arr[0:3, 0:3]
```

### Querying/Searching in an array

```python
horsepower = [130, 165, 150, 150, 140]
horsepower_arr=np.array(horsepower)

x = np.where(horsepower_arr >=150)
#gives the indices
print(x)
#with indices values can be found
horsepower_arr[x]
```

### Filtering Data

```python
horsepower = [130, 165, 150, 150, 140]
horsepower_arr=np.array(horsepower)

filter_arr = horsepower_arr>135
newarr = horsepower_arr[filter_arr]

print(filter_arr)
print(newarr)
```

### sorting an array

```python
horsepower = [130, 165, 150, 150, 140]
horsepower_arr=np.array(horsepower)

print('Original Array: ',horsepower_arr)
print('Sorted Array: ',np.sort(horsepower_arr))
print('Original Array after sorting : ',horsepower_arr)
```

```python
horsepower = [130, 165, 150, 150, 140]
horsepower_arr=np.array(horsepower)

print('Original Array: ',horsepower_arr)
horsepower_arr.sort()
print('Original Array after sorting : ',horsepower_arr)
```

array.sort(): modifies the original array by default <br>
sort(array): does not modify the original array by default

### Vectorized operations

Numpy makes use of optimized, pre-compiled code to perform mathematical operations on each array element. This eliminates the need of using loops, thereby enhancing the performance. This process is called vectorization.

```python
student_marks_arr=np.array([78, 92, 36, 64, 89])
print(np.sum(student_marks_arr))
```

```python
#award extra marks in each subject
additional_marks = [2,2,5,10,1]
student_marks_arr+=additional_marks
student_marks_arr
```

```python
student_marks_arr=np.array([78, 92, 36, 64, 89])
additional_marks = [2,2,5,10,1]
student_marks_arr = np.add(student_marks_arr,additional_marks)
student_marks_arr
```

### Broadcasting

```python
array1=np.array([5,10,15])
array2=np.array([5])
#array2 is being streched to match array1 resulting in new array with the same shape as array1
array3=array1*array2
array3
```

```python
#students marks in 4 subjects
students_marks = np.array([[67, 45],[90, 92],[66, 72],[32, 40]])
students_marks
```

```python
#adding 5 marks in 1st sub and 10 marks in 2nd sub
students_marks+=[5,10]
students_marks
```
