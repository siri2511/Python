### **Exercise** <br>
Display a message as follows for a given number:
1. multiple of 3: display "Zip"
2. multiple of 5: display "Zap"
3. multiple of both 3 and 5: display "Zoom"
4. no condition satified: display "Invalid"

```python
def func_z(n):
    if n%3==0 and n%5==0:
        print("Zoom")
    elif n%3==0:
        print("Zip")
    elif n%5==0:
        print("Zap")
    else:
        print("Invalid")
```

```python
n = int(input("Enter number:"))

func_z(n)
```

### **NESTED LOOP**

```python
number_of_passengers=5
number_of_baggage=2
security_check=True
for passenger_count in range(1,number_of_passengers+1):
    for baggage_count in range(1,number_of_baggage+1):
        if security_check:
            print("Security check of passenger:",passenger_count,"-- baggage:",baggage_count,"baggage_cleared")
        else:
            print("Security check of passenger:",passenger_count,"-- baggage:",baggage_count,"baggage_not_cleared")
```

```python
number_of_passengers=5
number_of_baggage=2
security_check=True
for passenger_count in range(1,number_of_passengers+1):
    baggage_count=1
    while baggage_count<=number_of_baggage:
        if security_check:
            print("Security check of passenger:",passenger_count,"-- baggage:",baggage_count,"baggage_cleared")
        else:
            print("Security check of passenger:",passenger_count,"-- baggage:",baggage_count,"baggage_not_cleared")
        baggage_count+=1
```

### **Exercise** <br>
Find the sum of digits of a given number

```python
def digits_sum(n):
    sum=0
    while n!=0:
        sum+=n%10
        n=n//10
    print("The sum is ",sum)
```

```python
n = int(input("Enter number "))

digits_sum(n)
```

### **Exercise** <br>
1. check if the given year is leap or not
2. find and display the maximum of given 3 numbers
3. check whether the given number is prime or not
4. print first n Fibonnaci numbers

```python
def check_leap(n):
    return (n%4==0) and ((n%100!=0)or(n%400==0))

n = int(input("Enter year:"))

if check_leap(n):
    print(n," is a leap year")
else:
    print(n," is not a leap year")
```

```python
def max_digit(a,b,c):
    if a>b and a>c:
        print(a," is maximum")
    if b>c:
        print(b," is maximum")
    else:
        print(c," is maximum")

a,b,c = (int(i) for i in input("Enter three numbers").split(","))
max_digit(a,b,c)
```

```python
import math
def check_prime(n,i):
    for j in range(2,i+1):
        if n%j==0:
            return False
    return True

n = int(input("Enter number:"))
if check_prime(n,int(math.sqrt(n))):
    print(n," is a prime")
else:
    print(n," is not a prime")
```

```python
def gen_fibonnaci(n):
    series = [0]*n
    series[0]=a=0
    series[1]=b=1
    if n > 2:
        for i in range(2,n):
            series[i] = series[i-1]+series[i-2]
    return series

n = int(input("Enter n value:"))
print(gen_fibonnaci(n))
```
