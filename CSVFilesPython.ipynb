## **PYTHON PROGRAMMING USING CSV FILES**

### **Reading a CSV file using reader()**

```python
import csv
with open(r'customers.csv','r') as csvfile:
    reader = csv.reader(csvfile)
    for record in reader:
        print(record)
```

### **Reading a CSV file using DictReader()**

```python
import csv
with open(r'customers.csv','r') as csvfile:
    reader = csv.DictReader(csvfile)
    for record in reader:
        print(record)
```

```python
#incase header row doesnt exist, then field naes can be passed as list argument to DictReader()
import csv
with open(r'customers.csv','r') as csvfile:
    fields=['GeographyKey','CustomerName','BirthDate','Mstatus','Gender','EmailAddress','YearlyIncome','Education','Occupation','NumberCarsOwned','Address','AddressLine1','PhoneNumber','DateFirstPurchase']
    reader=csv.DictReader(csvfile,fields)
    for record in reader:
        print(record)
```

### **Writing into CSV file using writer()**

```python
import csv
with open(r'employees.csv','w') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['EmpId','Emp Name','Salary'])
    writer.writerow([1,'Ramesh',22001.00])
    writer.writerow([2,'Rakesh',26501.00])
```

### **Writing into CSV file using DictWriter()**

```python
import csv
with open(r'employees.csv','w',newline='') as csvfile:
    fields=['EmpId','Emp Name','Salary']
    writer = csv.DictWriter(csvfile, fields)
    writer.writeheader()
    writer.writerow({'EmpId':1,'Emp Name':'Ramesh','Salary':22001.00})
    writer.writerow({'EmpId':2,'Emp Name':'Rakesh','Salary':26501.00})
```

```python
#incase of missing key-value pair, empty data gets inserted into the file
import csv
with open(r'employees.csv','a',newline='') as csvfile:
    fields=['EmpId','Emp Name','Salary']
    writer = csv.DictWriter(csvfile, fields)
    writer.writerow({'EmpId':3,'Salary':15000.00})
```
