## **PYTHON PROGRAMMING USING EXCEL FILES**

```python
import openpyxl as xl
```

```python
wb = xl.load_workbook('Employees.xlsx')
print('Type of wb:',type(wb))

print("Active worksheet object: ",wb.active)
print("Worksheet names: ",wb.sheetnames)
print('Worksheet objects: ',wb.worksheets)

ws=wb['Emp']
print('Type of ws:',type(ws))
```

```python
cell = work_sheet['A2']
print(cell.value)
```

### **Worksheet Slicing**

```python
ws = wb['Emp']

cells=ws['A1':'C4']

for row in cells:
    for col in row:
        print(col.value, end=' ')
    print()
```

### **Fetching all rows and columns values**

```python
wb = xl.load_workbook('Employees.xlsx')
ws=wb['Emp']

print(ws.max_row)
print(ws.max_column)
```

```python
for row in range(1,ws.max_row+1):
    for col in range(1,ws.max_column+1):
        print(ws.cell(row,col).value,end=' ')
    print()
```

### **Accessing sheet data row-wise and column-wise**

```python
wb = xl.load_workbook('Employees.xlsx')
ws = wb['Emp']

print('Accessing cell values row-wise: \n')
all_rows = tuple(ws.rows)
for rows in all_rows:
    for col in rows:
        print(col.value, end=' ')
    print()

print('Accessing cell values column-wise: \n')
all_columns = tuple(ws.columns)
for cols in all_columns:
    for row in cols:
        print(row.value, end=' ')
    print()
```

### **Accessing the entire sheet values**

```python
wb = xl.load_workbook('Employees.xlsx')
ws = wb['Emp']
for row in ws.values:
    for value in row:
        print(value, end=' ')
    print()
```

### **Fetching values from a matching row**

```python
wb = xl.load_workbook(r'Employees.xlsx')
ws=wb['Emp']

emp_id_to_match=3
flag=0

for row in range(2,ws.max_row+1):
    if ws.cell(row,1).value==emp_id_to_match:
        cells=ws[row]
        flag=1
        break

if flag:
    for cell in cells:
        print(cell.value, end=' ')
else:
    print('Employee detail does not exist')
```

### **Fetching values from a matching column**

```python
wb=xl.load_workbook(r'Employees.xlsx')
ws=wb['Emp']

dept_to_match='ETA'

cells=[]
for col in range(1,ws.max_column+1):
    if ws.cell(1,col).value=='Dept':
        for row in range(2,ws.max_row+1):
            if ws.cell(row,col).value==dept_to_match:
                cells.append(ws['B'+str(row)].value)
        break

if cells:
    for cell in cells:
        print(cell)
else:
    print('No Employees exist in '+dept_to_match+' department')
```

### **Adding a new record**

```python
wb=xl.load_workbook(r'Employees.xlsx')
ws=wb['Emp']

#Adding as a list
ws.append([4,'Payal',10000])

#Adding as a tuple
ws.append((5,'David',10000))

wb.save(r'Employees.xlsx')
```

### **Updating value for a specific cell**

```python
wb=xl.load_workbook(r'Employees.xlsx')
ws=wb['Emp']

ws['B4']='James'

emp_id_to_update = 2

for row in range(2,ws.max_row+1):
    if ws.cell(row,1).value==emp_id_to_update:
        ws.cell(row,3).value==50000
        break
else:
    print("Employee not found")

# any changes made will be persisted only if save method is invoked, else, changes will be lost
wb.save(r'Employees.xlsx')
```

### **Insert rows and columns**

```python
wb=xl.load_workbook(r'Employees.xlsx')
ws=wb['Emp']

ws.insert_rows(3)

#insert 2 rows before 5th row
ws.insert_rows(5,2)

wb.save(r'Employees.xlsx')
```

```python
wb=xl.load_workbook(r'Employees.xlsx')
ws=wb['Emp']

ws.insert_cols(2)

#inserts two columns before 4th column
ws.insert_cols(4,2)

wb.save(r'Employees.xlsx')
```

### **Deleting rows and columns**

```python
wb=xl.load_workbook(r'Employees.xlsx')
ws=wb['Emp']

#deletes 3rd row
ws.delete_rows(3)

#deletes two rows from 5th row
ws.delete_rows(5,2)

wb.save(r'Employees.xlsx')
```

```python
wb=xl.load_workbook(r'Employees.xlsx')
ws=wb['Emp']

#deletes 2n column
ws.delete_cols(2)

#deletes two columns from 3rd column
ws.delete_cols(3,2)

wb.save(r'Employees.xlsx')
```

```python
wb=xl.Workbook()
print(wb.sheetnames)
wb.create_sheet(title='Emp',index=0)
print(wb.sheetnames)
wb.remove(wb['Sheet'])
print(wb.sheetnames)
ws=wb['Emp']

ws['A1']='Name'
ws['B1']='Salary'
ws.append(['Ramu',23200.0])
ws.append(['Rakesh',21229.12])
wb.save(r'Employee1.xlsx')
```
